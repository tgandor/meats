# Applying functions returning Series

on dataframes and groupbys

```python
import pandas as pd
```

```python
pd.Series({'a': 1, 'b': 2})
```

```python
def konst(g):
    # print('konst arg:', g)
    return pd.Series({'a': 1, 'b': 2})

df = pd.DataFrame.from_dict({
    1: 3,
    2: 4,
    3: 10
}, orient='index')
df
```

```python
df.apply(konst)
```

```python
df.apply(konst, axis=0)
```

```python
df.apply(konst, axis=1)
```

```python
df.groupby(0).apply(konst)
```
